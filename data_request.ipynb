# I requested the data available on the Committee to Protect Journalists (CPJ) website.
_There was no way to download a spreadsheet directly from CPJ data, so I found an API "hidden" on the website._

```python
import requests
import pandas as pd

base_url = "https://cpj.org/wp-json/cpj-datamanager/v1/people_list"
params = {
    "status": "Killed",
    "motive": "Confirmed",
    "role": "Journalist",
    "sortBy": "fullName",
    "page": 1
}

all_data = []
```

```python
while True:
    print(f"Collecting page {params['page']}...")
    response = requests.get(base_url, params=params)
    data = response.json()

    # Verifica se veio vazio, para sair do loop
    if not data['data']:
        print("No more data. Ending loop.")
        break

    for person in data['data']:
        row = {
            "Nome": person.get("fullName"),
            "Data da Morte": person.get("startDisplay"),
            "Local da Morte": person.get("location"),
            "Organizações": person.get("organizations"),
            "Tipo": person.get("type"),
            "Causa": person.get("typeOfDeath"),
            "Motivo Confirmado": person.get("motiveConfirmed"),
            "Link": person.get("mtpage")
        }
        all_data.append(row)

    params["page"] += 1

#Save Data
df = pd.DataFrame(all_data)
print(df.head())
df.to_excel("jornalists_murdered.xlsx", index=False)
```
